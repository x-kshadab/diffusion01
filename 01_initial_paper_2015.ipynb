```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import torch
from sklearn.datasets import make_swiss_roll
```

```python
n_samples=5000
swiss_role_data, _= make_swiss_roll(n_samples)
swiss_role_data.dtype, swiss_role_data.shape,swiss_role_data[:5]
```

```python
# we dont need 3rd dimesnion of data as we want a 2d figure
swiss_role_data_org = swiss_role_data[:,[0,2]]
# flip
swiss_role_data_1 = swiss_role_data[:,[2,0]]/10
#to rotate the swiss roll into anti-clockwise direction
swiss_role_data_2=swiss_role_data_1 * np.array([1,-1])
# swiss_role_data.shape, swiss_role_data[:5]
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.scatter(swiss_role_data_org[:,0], swiss_role_data_org[:,1])
plt.subplot(1,3,2)
plt.scatter(swiss_role_data_1[:,0], swiss_role_data_1[:,1])
plt.subplot(1,3,3)
plt.scatter(swiss_role_data_2[:,0], swiss_role_data_2[:,1])
plt.xlim([-2,2])
plt.ylim([-2,2])
```
